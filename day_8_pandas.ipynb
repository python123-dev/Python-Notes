```python
# Create Function
def times1000(x):
    return format(x * 1000, ",d")
```

```python
my_df['Salary'].apply(times1000)
```

```python
# Add Elder Last Name
def namer(x):
    if x == "John":
        return "John Elder"
    else:
        return "Not Important"
```

```python
# Show Elder Last Name
my_df["Employees"].apply(namer)
```

```python
# Use Lambda
pd.DataFrame(my_df["Salary"].apply(lambda x: format(x * 1000, ",d")))
```

```python
# Apply To Two Columns
my_df[["Corporation", "Employees"]].apply(namer)
```
