```
import matplotlib.pyplot as plt

values=[5,8,9,10,4,7]
colors=['b','g','r','c','m','y']
labels=['A','B','C','D','E','F']
explode=(0,0.2,0,0,0,0)

plt.pie(values,colors=colors,labels=labels,explode=explode,shadow=True,counterclock=False,autopct='%1.1f%%')

plt.show()
```

```
import pandas_datareader as web
import matplotlib.pyplot as plt

import mplfinance as mpf

import datetime as dt
```

```
company="^NSEI"
#currency="USD"
start=dt.datetime(2020,1,1)
end=dt.datetime.now()
data=web.DataReader(company,"yahoo",start,end)
```

```
print(data)
```

```
plt.plot(data['Close'])
plt.show()
```

```
mpf.plot(data,type="candle")
```

```
mpf.plot(data,type="candle",volume=True,style="yahoo")
```

```
data1=web.DataReader("AAPL","yahoo",start,end)
data2=web.DataReader("TSLA","yahoo",start,end)
```

```
plt.plot(data1["Close"],label="AAPL")
plt.plot(data2["Close"],label="TSLA")
plt.legend(loc="upper left")
plt.show()
```

```
plt.yscale("log")
plt.plot(data1["Close"],label="AAPL")
plt.plot(data2["Close"],label="TSLA")
plt.legend(loc="upper left")
plt.show()
```
